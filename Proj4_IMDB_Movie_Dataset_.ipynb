```python
#20. Classify Movies Based on Ratings [Excellent, Good, and Average]
def rating(rate):
  if rate > 8:
    return "Excellent"
  elif 6 < rate >=8 :
    return "Good"
  else:
    return "Average"
```

```python
#22. Find Unique Values From Genre
movie['Genre']
```

```python
#step1 spliting the entire list using ','
lst1=[]
for val in movie['Genre']:
  lst1.append(val.split(','))
```

```python
# lst1 obitaned is a 2D list and we convert the whole list into a 1D list to find the unique values
one_D=[]
for item in lst1:
  for item1 in item:
    one_D.append(item1)
```

```python
one_D
```

```python
# counting unique value from list
unique_lst =[]
for itm in one_D:
  if itm not in unique_lst:
    unique_lst.append(itm)
```
